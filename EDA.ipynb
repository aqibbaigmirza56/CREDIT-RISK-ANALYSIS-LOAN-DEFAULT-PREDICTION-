```python
## step 3 : EDA 


import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df=pd.read_csv("Loan_default.csv")
df.head()
```

```python
sns.countplot(x='Default',data=df)
plt.title("LOAN DEFAULT DISTRIBUTION")
plt.show()
```

```python

sns.histplot(df['Age'],kde=True,edgecolor="black")
plt.title("AGE DISTRIBUTION")
plt.show()
```

```python
sns.histplot(df["LoanAmount"], kde=True)
plt.title("Loan Amount Distribution")
plt.show()
```

```python
sns.histplot(df['Income'], kde=True)
plt.title("INCOME DISTRIBUTION")
plt.show()
```

```python
sns.histplot(df["CreditScore"], kde=True)
plt.title("Credit Score Distribution")
plt.show()
```

```python
sns.boxplot(x="Default", y="Income", data=df)
plt.title("Income vs Loan Default")
plt.show()
```

```python
sns.boxplot(x="Default", y="CreditScore", data=df)
plt.title("Credit Score vs Loan Default")
plt.show()
```

```python
sns.boxplot(x="Default", y="DTIRatio", data=df)
plt.title("DTI Ratio vs Loan Default")
plt.show()
```

```python
sns.countplot(x="EmploymentType", hue="Default", data=df)
plt.title("Employment Type vs Loan Default")
plt.show()
```

```python
sns.countplot(x="LoanPurpose", hue="Default", data=df)
plt.title("Loan Purpose vs Loan Default")
plt.show()
```

```python
print("This bar chart compares the average credit score of defaulters and non-defaulters.\n It clearly shows that customers who default have a lower average credit score, highlighting credit score as a key risk factor in loan default prediction")

sns.barplot(
    x="Default",
    y="CreditScore",
    data=df,
    width=0.1,
    hue="Default",
    palette=["steelblue", "orchid"]
)

plt.title("Credit Risk Analysis: Average Credit Score by Loan Default Status")
plt.xlabel("Loan Default (0 = No, 1 = Yes)")
plt.ylabel("Average Credit Score")

plt.legend(title="Default Status", loc="upper right")
plt.show()s
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn theme
sns.set_theme(
    style="darkgrid",
    context="notebook",
    font_scale=1.1
)

# Custom color palette 
custom_palette = [
    "#6A5ACD",  # slate blue
    "#00B3B3",  # teal
    "#FF8C69",  # coral
    "#8B008B",  # dark magenta
    "#4682B4",  # steel blue
    "#2F4F4F",  # dark slate gray
    "#9370DB",  # medium purple
    "#20B2AA"   # light sea green 
]

sns.set_palette(custom_palette)

# Matplotlib tweaks
plt.rcParams["figure.figsize"] = (8, 5)
plt.rcParams["axes.titlesize"] = 14
plt.rcParams["axes.labelsize"] = 12
```

```python
sns.histplot(
    df['CreditScore'],
    kde=True,
    color="#6A5ACD"
)
plt.title("CREDIT SCORE DISTRIBUTION", fontweight="bold")
plt.show()
```

```python
sns.boxplot(
    x='Default',
    y='Income',
    data=df,
    palette=["#8B008B", "#FF8C69"]
)
plt.title("Income vs Loan Default", fontweight="bold")
plt.show()
```

```python
sns.boxplot(
    x='Default',
    y='CreditScore',
    hue='Default',
    data=df,
    dodge=False
)
plt.title("Credit Score vs Loan Default")
plt.show()
```
